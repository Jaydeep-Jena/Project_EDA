```python
top_5_counts = df_chennai['name'].value_counts().head(5)
top_5_df = pd.DataFrame({'Value': top_5_counts.index, 'Count': top_5_counts.values})
```

```python
plt.figure(figsize=(15,4))
plt.bar(top_5_df['Value'],top_5_df['Count'])
plt.show()
```
